```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
data = pd.read_csv('/gdrive/Shared drives/KPU_project/climate-energy.csv')
data
```

```
data01 = data.drop(0, axis = 0)
data01
```

```
dataTestY = data01.values[:,3]
dataTestY01 = dataTestY.astype(float)
dataTestY02 = dataTestY01.astype(int)
dataTestY02
```

```
userPeakValue = 9000 # 사용자가 원하는 전력 제한 피크 값
testY = dataTestY02//userPeakValue
testY.shape
```

```
dataTestX=data01[['EMXT',	'EMNT',	'MMXT',	'MMNT',	'MNTM']]
testX = dataTestX.values
testX
```

```
from sklearn.model_selection import train_test_split
```

```
def 모델평가(model, X, y, **설정):
    X_train, X_test, y_train, y_test = train_test_split(X, y, **설정)
    model.fit(X_train, y_train)
    훈련점수 = model.score(X_train, y_train)
    시험점수 = model.score(X_test, y_test)
    return 훈련점수, 시험점수
```

```
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
```

```
tree_clf = DecisionTreeClassifier(max_depth=4)
모델평가(tree_clf, testX, testY, random_state=2)
```

```
특성중요도 = tree_clf.feature_importances_
```

```
np.sum(특성중요도)
```

```
pd.Series(특성중요도, index=['EMXT',	'EMNT',	'MMXT',	'MMNT',	'MNTM']).plot(kind='barh') #정확도에 영향을 주는 순서
```

```
tree_clf = DecisionTreeRegressor(max_depth=10)
모델평가(tree_clf, testX, testY, random_state=1)
#DecisionTreeRegressor 못써먹을 듯
```
