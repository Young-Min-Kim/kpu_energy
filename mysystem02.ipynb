```
data = pd.read_csv('/gdrive/My Drive/Colab Notebooks/climate-energy.csv')
data
```

```
data01 = data.drop(0, axis = 0)
data01
```

```
dataTestY = data01.values[:,3]
dataTestY
dataTestY01 = dataTestY.astype(float)
dataTestY01
dataTestY02 = dataTestY01.astype(int)
dataTestY02
testY = dataTestY02//1000
testY.shape
```

```
dataTestX=data01[['EMXT',	'EMNT',	'MMXT',	'MMNT',	'MNTM']]
testX = dataTestX.values
testX
```

```
from sklearn.model_selection import train_test_split
```

```
def 모델평가(model, X, y, **설정):
    X_train, X_test, y_train, y_test = train_test_split(X, y, **설정)
    model.fit(X_train, y_train)
    훈련점수 = model.score(X_train, y_train)
    시험점수 = model.score(X_test, y_test)
    return 훈련점수, 시험점수
```

```
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
```

```
tree_clf = DecisionTreeClassifier(max_depth=2)
모델평가(tree_clf, testX, testY, random_state=3)
```

```
tree_clf.tree_.max_depth
```

```
from sklearn.tree import plot_tree
```

```
그림틀 = plt.figure(figsize=(20, 10))
그래프 = plot_tree(tree_clf)
plt.savefig('cli-eng tree.png')
```

```
특성중요도 = tree_clf.feature_importances_
```

```
np.sum(특성중요도)
```

```
pd.Series(특성중요도, index=['EMXT',	'EMNT',	'MMXT',	'MMNT',	'MNTM']).plot(kind='barh') #정확도에 영향을 주는 순서
```
