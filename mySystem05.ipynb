# 이후 전력 사용량 예측
기존의 피크 데이터와 함께 사용할 전력 사용량 예측

```
from sklearn.model_selection import train_test_split
```

```
def 모델평가(model, X, y, **설정):
    X_train, X_test, y_train, y_test = train_test_split(X, y, **설정)
    model.fit(X_train, y_train)
    훈련점수 = model.score(X_train, y_train)
    시험점수 = model.score(X_test, y_test)
    return 훈련점수, 시험점수
```

```
from sklearn.neighbors import KNeighborsRegressor
```

```
# testY 단위를 100단위로 설정해서 정확도는 매우 낮음/ 예측 모델이 얼마나 기존 데이터에 근접하는지가 주안점
energeModel_Neighbor = RandomForestClassifier(n_estimators=100)
모델평가(energeModel_Neighbor, testX, testY_ValuePredict)
```

```
#임의의 데이터가 들어왔을 때 해당 데이터의 예측 값을 출력해주는 코드
test_features = [[29.97,  6.77, 26.11, 14.18, 20.15]] # 67의 데이터를 70으로 예측 해냄
pred_forest = energeModel_Neighbor.predict(test_features)
print(pred_forest)
```

미래의 전력량 예측을 위해서는 머신러닝으로 학습한 기존의 시간 데이터(시간과 전력 량의 상관 관계)
온습도(전날과 상이함) -> 기상청의 시간 단위 다음 날의 날씨 예보를 참고
등을 파악하여 다음 날의 시간 당 전력 사용량 그래프를 출력 해줄 수 있을 것
